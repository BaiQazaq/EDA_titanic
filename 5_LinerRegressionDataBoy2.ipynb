```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('CSV/bitcoin.csv')
data.head()
```

```python
#Удаляем колонку Unnamed:0
data.drop('Unnamed: 0', axis=1, inplace=True)
data
```

```python
projection = 14
data['Predict'] = data['close'].shift(-projection)
data
```

```python
x = data[['close']]
y = data['Predict']
```

```python
x.shape
```

```python
X = x[:-projection]
Y = y[:-projection]
```

```python
#Separate the data to "train" and "test" dataset test_size = 0.2 it mean 20% will for testing
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size= 0.2, random_state=42)  
```

```python
model1 = LinearRegression()   #Creating model
```

```python
model1.fit(x_train, y_train) #Training model1 by data set x_train y_train
```

```python
f"Точность модели составляет = {model1.score(x_test, y_test)}"
```
